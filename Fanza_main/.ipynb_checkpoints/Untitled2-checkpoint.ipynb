```python
def soup2(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    res2 = requests.get(soup.find('a', class_ = "ageCheck__link ageCheck__link--r18")['href'])

    soup2 = BeautifulSoup(res2.text, 'html.parser')
    return soup2
```
