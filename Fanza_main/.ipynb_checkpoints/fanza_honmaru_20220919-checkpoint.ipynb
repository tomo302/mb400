```python
def actor_max_page(soup2):
    #５０音キーワード別女優総人数：total_number_of_actor
    total_number_of_actor= soup2.find('div', class_ = "list-boxcaptside list-boxpagenation group").p.text.replace('　', '').split('人')[0]
    max_page_number = soup2.find('div', class_ = "list-boxcaptside list-boxpagenation group").p.text.replace('　', '').split('全')[1].split('ペ')[0]
    return max_page_number
```

```python
def genre_sakuhi_maxpage(soup2):
    #ジャンルページは、「10,581タイトル中　1～120タイトル　1ページ目を表示」で全ページ数の表示がないのでタイトル数//１２０で切り上げ割り算でページ数を取得する
    #.replace(',', '')はタイトル数が４桁になると「1,000」とカンマが入るので削除している。int()に変換できないので。この記述がタイトル数３桁でエラーを発生することはない。
    sakuhin_title_number = int(soup2.find('div', class_ ="list-boxcaptside list-boxpagenation").p.text.replace('　', '').split('タ')[0].replace(',', ''))
    # 121//120は切り捨て割り算、期待値は切り上げ割り算の結果で下記のようになる。
    max_page_number = -(-sakuhin_title_number // 120)
    return max_page_number
```

```python
def soup2(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    res2 = requests.get(soup.find('a', class_ = "ageCheck__link ageCheck__link--r18")['href'])

    soup2 = BeautifulSoup(res2.text, 'html.parser')
    if "d/i" in url:
        print("ジャンルページです")
        #ジャンルページ用の関数を実行
        max_page = genre_sakuhi_maxpage(soup2)
        url_kind = "genre"
        baseurl = "https://www.dmm.co.jp/digital/videoa/-/list/=/article=keyword/id=1001/page="
        return soup2, max_page, url_kind, baseurl
    elif "a/-/a" in url:
        print("５０音ページです")
        url_kind ="50on"
        max_page =  actor_max_page(soup2)
        actor_lists = actor_data(soup2)
        baseurl = "https://www.dmm.co.jp/digital/videoa/-/actress/=/keyword=a/page="
        #５０音ページ用の関数を実行
        return soup2, url_kind, max_page, baseurl, actor_lists 
    elif "t&" in url:
        print("ビデオページです")
        url_kind = "video"
        sakuhin_data = get_sakuhin_data(soup2)
        #ビデオページ用の関数を実行
        return sakuhin_data
    elif "s/i" in url:
        print("出演一覧ページ")
        url_kind = "syutuen_itiran"
        max_page = genre_sakuhi_maxpage(soup2)
        syutuen_data = syutuen_itiran(soup2)
        baseurl = "https://www.dmm.co.jp/digital/videoa/-/list/=/article=actress/id=1008887/page="
        return soup2, url_kind, max_page, baseurl, syutuen_data
        #出演一覧ページ用の関数を実行
    #return url_kind, max_page, baseurl 
```

```python
def get_sakuhin_data(soup2):
   
    
    #作品データ詳細取得：作品タイトル、作品内容詳細、作品内画像URL、DVDジャケット見開き画像URL、サンプル動画URL、出演者名
    json_2 = soup2.find("script", {"type": "application/ld+json"})
    data = json_2.text
    json_dict = None
    #サンプル動画の有無を確認
    if "subjectOf" in data and "contentUrl" in data:
        json_dict = json.loads(data)
        
        sakuhin_name = json_dict['name']
        sakuhin_description = json_dict['description']
        sakuhin_image = json_dict['image']
        dvd_jacket_url = soup2.find("div", class_ = "tx10 pd-3").a['href']
        sample_video_url = json_dict['subjectOf']['contentUrl']#デフォルットだと小さい画像なので
        try:
            actor_name = json_dict['subjectOf']['actor']['name']
        except KeyError:
            actor_name = "出演者複数"
            print(actor_name)
            
        #サンプル画像URLを取得してリストで保存    
        sample_gazou = []
        for _ in soup2.find_all("img", class_ ="mg-b6"):
            sample_gazou.append(_['src'])
        
    #取得してリターンする値：：作品タイトル、作品内容詳細、作品内画像URL、DVDジャケット見開き画像URL、サンプル動画URL、出演者名、作品サンプル画像
    sakuhin_lists = []
    sakuhin_lists.append([json_dict['name'], json_dict['description'], json_dict['image'], dvd_jacket_url, json_dict['subjectOf']['contentUrl'].replace('sm', 'mhb'), actor_name, sample_gazou])
    
    #return sakuhin_name, sakuhin_description, sakuhin_image, sample_video_url
    return sakuhin_lists
```

```python
def video(soup2):
    json_2 = soup2.find('script', type="application/ld+json")
    data = json_2.text
    json_dict = None
    
    sample_jpgs = []
    for row in soup2.find_all('img', class_ = 'mg-b6'):
        sample_jpg = (row['src'])
        sample_jpgs.append(sample_jpg)
    return sample_jpgs
```

```python
def actor_data(soup2):
    actor_data = actor.find("ul", class_ = "d-item act-box-100 group")
    act_lists = []
    for row in actor_data.select('li a'):
        actor_name = (row.img['alt'])
        furigana = (row.span.text)
        syutuensu = (row.span.next_sibling.text)
        actor_jpgurl = (row.img['src'])
        syutuen_itiran_page_url = (row['href'])
        act_list = [actor_name, furigana, syutuensu, actor_jpgurl, syutuen_itiran_page_url]
        act_lists.append(act_list)
    return act_lists
```

```python
def syutuen_itiran(soup2):
    syutuen_data = []
    actor_name = soup2.find_all("span", itemprop="name")[3].text
    for row in soup2.find_all('p', 'tmb'):
        video_title = (row.img['alt'])
        sakuhin_url = (row.a['href'])
        sakuhin_jpg_url = (row.img['src'])
        syutuen_list = [actor_name, video_title, sakuhin_url, sakuhin_jpg_url]
        syutuen_data.append(syutuen_list)
    
    return syutuen_data
        
```
